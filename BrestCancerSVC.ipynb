Imorting Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.svm import SVC
```

Reading CSV file as Pandas Data frame

```python
BrCan = pd.read_csv('cancer.csv')
BrCan.head(10)
```

```python
BrCan.dtypes
```

```python
pp = BrCan['diagnosis'].values
print(np.unique(pp))
```

Processing Lable Encoding For features 

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
PP = labelencoder.fit_transform(pp) # M = 1 and B = 0
print(np.unique(PP))
```

PairPlot to Review Data 

```python
sns.pairplot(BrCan,hue='diagnosis',vars = ['radius_mean', 'texture_mean', 'perimeter_mean','area_mean','smoothness_mean'])
plt.figure(figsize=(7,7))
```

```python
sns.heatmap(BrCan['radius_mean texture_mean perimeter_mean area_mean smoothness_mean'.split()].corr(), annot=True)
sns.scatterplot(x='texture_mean',y='perimeter_mean',hue= 'diagnosis',data = BrCan)
```

Removing Null Values

```python
BrCan.isna().sum()
BrCan = BrCan.dropna(axis = 1)
BrCan.shape
BrCan['diagnosis'].value_counts()
```

Count Plot B/w Malignan VS Benign cases

```python
sns.countplot(x='diagnosis',data=BrCan)
plt.title('Number of Malignan VS Benign')
```

Assigning Test and Train data

```python
x = BrCan.drop(['diagnosis'],axis=1).values
#x= BrCan['diagnosis'].values
y= BrCan['diagnosis'].values
from sklearn.model_selection import train_test_split
x_train, x_test ,y_train, y_test = train_test_split(x,y,test_size=0.10,random_state=44)
```

Fitting Train Test to SVC model

```python
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(x_train,y_train)
```

Heat Map of Confusion Matrix 

```python
from sklearn.metrics import classification_report , confusion_matrix
y_pred = svc_model.predict(x_test)
conmx = confusion_matrix(y_test,y_pred)
sns.heatmap(conmx, annot=True)
```

```python
#normalization

from sklearn.preprocessing import MinMaxScaler
mn_scaler = MinMaxScaler()
x_train_scaled = mn_scaler.fit_transform(x_train.astype(np.float_))
x_test_scaled = mn_scaler.fit_transform(x_test.astype(np.float_))

```

```python
#SVM
from sklearn.svm import SVC
svc_mod = SVC()
svc_mod.fit(x_train_scaled, y_train)
from sklearn.metrics import classification_report, confusion_matrix
y_predict_scaled = svc_model.predict(x_test_scaled)
cm = confusion_matrix(y_test, y_predict_scaled)
sns. heatmap(cm, annot=True)
print(classification_report(y_test, y_predict_scaled))
```
